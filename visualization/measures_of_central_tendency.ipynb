```python
import pandas as pd
df = pd.read_csv('../data/dummy_data/hotel_reviews.csv')

# 平均
df.mean()
df['満足度'].mean()

# 最頻値
# NOTE:最頻値は複数ある可能性があるのでシリーズ型で返される
df.mode()
df['満足度'].mode()

# 中央値
df.median()
df['満足度'].median()

# 個数
df['満足度'].value_counts()

# 割合に変換
df['満足度'].value_counts(normalize=True)

# 代表値を一気に算出
# NOTE: 引数include='all'を指定すると全ての列に対して代表値を算出する（指定していない場合は、数値データのみ扱う）
df.describe()
```
