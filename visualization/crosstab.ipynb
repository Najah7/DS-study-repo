```python
import pandas as pd

import japanize_matplotlib

import seaborn as sns

df = pd.read_csv('../data/dummy_data/hotel_reviews.csv')

import japanize_matplotlib

# クロス集計
crosstab = pd.crosstab(df['満足度'], df['性別'])

# クロス集計の可視化(ヒストグラム)
crosstab.plot.bar()

# seabornを使った可視化。
# クロス集計のヒストグラムで扱いやすく、見やすい。
# NOTE：hue 引数は、カテゴリ変数を分割するための列名を指定するための引数
sns.countplot(x='満足度', hue='性別', data=df)
```
